```python
import qmcpy as qp
import numpy as np

import matplotlib.pyplot as plt
import pandas as pd
```

```python
sobol_lattice = qp.Sobol(dimension = 2, 
                         seed = 20210702,
                        )
```

```python
uniform_transform = qp.Uniform(sobol_lattice, 
                               lower_bound=[0.54, 0.3e6], 
                               upper_bound=[2.7, 1.1e6])
```

```python
qmc_points = uniform_transform.gen_samples(256)
```

```python
qmc_points
```

```python
sobol2 = qp.Sobol(dimension = 3, 
                         seed = 20210702,
                        )
```

```python
sobol_lattice.gen_samples(4)
```

```python
sobol2.gen_samples(4)
```

```python
qmc_pts_2 = uniform_transform.gen_samples(4)
qmc_pts_2
```

```python
def plot_qmc(qmc_pts_2d_arr):
    plt.scatter(qmc_pts_2d_arr[:, 0], qmc_pts_2d_arr[:, 1])
```

```python
plot_qmc(qmc_pts_2)
```

```python
qmc_pts_3 = uniform_transform.gen_samples(8)
qmc_pts_3
```

```python
maxpro_points = np.loadtxt("./MaxPro_BrFPF_256.csv", delimiter=',', skiprows=1, usecols=(1, 2))
```

```python
maxpro_points
```

```python
plt.scatter(qmc_points[:, 0], qmc_points[:, 1])
```

```python
plt.scatter(maxpro_points[:, 0], maxpro_points[:, 1], c="red")
```

```python
import seaborn as sns

p1 = sns.jointplot(x=qmc_points[:, 0], y=qmc_points[:, 1], kind='scatter')
p1.fig.suptitle("Sobol Points")
p1.ax_joint.collections[0].set_alpha(0.5)
p1.fig.tight_layout()
p1.fig.subplots_adjust(top=0.95)
p1.set_axis_labels('BrFactor', 'PoyntingFlux', fontsize=16)
```

```python
p2 = sns.jointplot(x=maxpro_points[:, 0], y=maxpro_points[:, 1], kind='scatter', color='red')
p2.fig.suptitle("MaxPro Points")
p2.ax_joint.collections[0].set_alpha(0.5)
p2.fig.tight_layout()
p2.fig.subplots_adjust(top=0.95)
p2.set_axis_labels('BrFactor', 'PoyntingFlux', fontsize=16)
```
